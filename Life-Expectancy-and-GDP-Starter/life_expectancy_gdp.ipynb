```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
life_gdp = pd.read_csv('all_data.csv')
print(life_gdp.head())
```

```python
print(life_gdp.Country.unique())
print(life_gdp.Year.unique())
```

```python
life_gdp = life_gdp.rename(columns={'Life expectancy at birth (years)': 'Life'})
print(life_gdp.head())
```

```python
print(life_gdp.describe(include='all'))
```

```python
sns.scatterplot(x='Life', y='GDP', data=life_gdp, hue='Country')
plt.show()
plt.clf()
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)
fig.tight_layout()

fig.subplots_adjust(wspace=0.5, hspace=0.5)

ax1.scatter(life_gdp.Life[life_gdp.Country=='Chile'], life_gdp.GDP[life_gdp.Country=='Chile'])
ax1.set_title('Chile')
ax2.scatter(life_gdp.Life[life_gdp.Country=='China'], life_gdp.GDP[life_gdp.Country=='China'])
ax2.set_title('China')
ax3.scatter(life_gdp.Life[life_gdp.Country=='Germany'], life_gdp.GDP[life_gdp.Country=='Germany'])
ax3.set_title('Germany')
ax4.scatter(life_gdp.Life[life_gdp.Country=='Mexico'], life_gdp.GDP[life_gdp.Country=='Mexico'])
ax4.set_title('Mexico')
ax5.scatter(life_gdp.Life[life_gdp.Country=='United States of America'], life_gdp.GDP[life_gdp.Country=='United States of America'])
ax5.set_title('United States of America')
ax6.scatter(life_gdp.Life[life_gdp.Country=='Zimbabwe'], life_gdp.GDP[life_gdp.Country=='Zimbabwe'])
ax6.set_title('Zimbabwe')
```

```python
print('Chile avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='Chile'].mean())
print('China avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='China'].mean())
print('Germany avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='Germany'].mean())
print('Mexico avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='Mexico'].mean())
print('United States of America avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='United States of America'].mean())
print('Zimbabwe avg life expectancy')
print(life_gdp.Life[life_gdp.Country=='Zimbabwe'].mean())
```
